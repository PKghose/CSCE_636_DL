```python
import tensorflow as tf
import matplotlib.pyplot as plt
```

```python
x = tf.Variable([0.0, 0.1, 0.2, 0.3], dtype=tf.float32)
with tf.GradientTape() as tape:
    f_x = tf.sin(x)
gradient = tape.gradient(f_x, x)

print("x:", x.numpy())
print("sin(x):", f_x.numpy())
print("gradient:", gradient.numpy())
```

```python
plt.figure(figsize=(10, 6))
plt.plot(
    x, gradient,
    label='∇sin(x)',
    color='green',
    linewidth=3,
    linestyle='-',
    marker='o',
    markerfacecolor='blue',
    markersize=7
)
plt.title('Gradient of f(x) = sin(x) with Respect to x')
plt.xlabel('x values')
plt.ylabel('Gradient ∂f/∂x')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
